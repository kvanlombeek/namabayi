```python
i = 0
#found_names = []
def handle_request(response):
    global i
    global found_names
    print('i = %i , response code = %i '%(i,response.code))
    if(response.code != 200):
        i -= 1
        return None
    request_bs = BeautifulSoup(response.body,"lxml" )
    links = request_bs.findAll('a')
    for link in links:
        # Check if the link matches the name pattern
        if(re.match(pattern = '/names/', string = link['href'])):
            found_names.append(link['href'])
    # Remove duplicates already found
    found_names = list(set(found_names))
    # Write to csv
    pd.Series(found_names).to_csv('scraped_data/found_names.csv')
    i -= 1

http_client = httpclient.AsyncHTTPClient(max_clients = 5)
for page_number in range(50, 100):
    url = 'http://babynames.net/names?page='+str(page_number)
    i += 1    
    http_client.fetch(url.strip(), handle_request)
```
